```python
########################################################
# Index & Unstack works like a pivot table in Excel
########################################################

# Indexing: Moves a column to a dimension: index is for rows
# Unstack: Moves an index into a column: unstack is for columns
```

```python
import pandas as pd
```

```python
movies = pd.read_csv('imdbratings.csv') # pd.read_csv('http://bit.ly/imdbratings')
```

```python
# Sort by index
movies.head()
```

```python
m = movies.set_index(['title'])
m.head()
```

```python
# Index work better and quicker when you sort the index
m = movies.set_index(['title']).sort_index()
```

```python
m = movies.set_index(['title', 'content_rating']).sort_index()
m
```

Time it takes to run function

```python
%%time
m.loc['The Godfather']
```

```python
# Reset an index
m.loc['The Godfather'].reset_index(['content_rating'])
```
