```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
# example of Single Aggregation after groupby
drinks.groupby('continent').beer_servings.mean()
```

```python
# multiple aggrgation functions can be applied simultaneously
drinks.groupby('continent').beer_servings.agg(['mean','min','max'])
```

```python
# new apply to a series
drinks.beer_servings.agg(['mean','min','max'])
```

```python
drinks.agg(['mean','min','max'])
```
